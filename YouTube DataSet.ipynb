```python
#  Reading the data
df = pd.read_csv('MrBeast_youtube_stats.csv')
```

```python
df = df.iloc[:,:8]
```

```python
df.shape
```

```python
# Subsetting the data with different metthods
df_subset_1 = df[df["viewCount"] > 1_000_000]
df_subset_2 = df.loc[df["viewCount"] > 1_000_000]
df_subset_3 = df.query("viewCount > 1000000")
```

```python
# how to request the oppesite of what you want by ~
df = df[~df.viewCount.isna()]
```

```python
# Casting dType
df.dtypes
df['viewCount'] = df["viewCount"].astype('int')
```

```python
df['publishTime'] = pd.to_datetime(df['publishTime'])
```

```python
# Creating new Columns
df['Ratio_view_like'] = df['likeCount'] / df['viewCount']
```

```python
# Adding new Row
df_to_append = df.tail(1)
df_concat = pd.concat([df, df_to_append])
df_concat
```
