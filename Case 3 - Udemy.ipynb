```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# here i convert the published_timestamp column to datetime because it was object 
data = pd.read_csv('udemy_courses.csv', parse_dates=['published_timestamp'])
```

```python
data = data.drop_duplicates()
```

```python
data.head()
```

```python
# Find out Number of course per subjects
data['subject'].value_counts()
```

```python
sns.countplot(data['subject'])
plt.xticks(rotation=45)
plt.show()
```

```python
# Which levels, udemy courses providing the course
data.level.value_counts()
```

```python
sns.countplot(data['level'])
plt.xticks(rotation=45)
plt.show()
```

```python
data.head()
```

```python
# Display the count of paid and free courses
data.is_paid.value_counts()
```

```python
print(f"The Total Free Course : {data.is_paid.value_counts()[0]}")
print(f"The Total Paid Course : {data.is_paid.value_counts()[1]}")
```

```python
# Which Course has more lectures (free or paid) ?
data.groupby('is_paid').sum()
```

```python
data.groupby('is_paid').mean()
```

```python
# Which Courses have a higher number of subscribers free or paid
sns.barplot(x='is_paid', y='num_subscribers', data=data)
```

```python
# Which level has high subscribers
sns.barplot(x='level', y='num_subscribers', data=data)
plt.xticks(rotation=45)
plt.show()
```

```python
# Find most popular course title
[data.groupby('course_title')['num_subscribers'].mean().sort_values(ascending=False).index[0]]
```

```python
data.loc[data['num_subscribers'].max() == data['num_subscribers']]
```

```python
# Display most 10 popular courses by sub
data.groupby('course_title')['num_subscribers'].mean().sort_values(ascending=False).head(10)
```

```python
top10 = data.sort_values(by='num_subscribers', ascending=False).head(10)
top10
```

```python
# Which level has high subscribers
sns.barplot(x='num_subscribers', y='course_title', data=top10)
plt.xticks(rotation=45)
plt.show()
```

```python
# Find the course which has the highest reviews
data.sort_values(by='num_reviews', ascending=False).head(1)
```

```python
# Does price affect number of reviews
data.corr()
```

```python
plt.figure(figsize=(15,6))
sns.scatterplot(x='price', y='num_reviews', data=data)  # Yes affect
```

```python
# Total Courses related to python
data['course_title'].str.contains('Python').sum()
```

```python
# Most 10 Popular course for Python (Sub)
python_10 = data.loc[data['course_title'].str.contains('Python')].sort_values(by='num_subscribers', ascending=False).head(10)
python_10
```

```python
sns.barplot(x='num_subscribers', y='course_title', data=python_10)
plt.xticks(rotation=45)
plt.show()
```

```python
# In Which year the highest number of courses were posted
data['year'] = data['published_timestamp'].dt.year
```

```python
data.groupby('year')['course_title'].count()
```

```python
sns.countplot('year', data=data)
```

```python
# Display category-wise count of posted subjects (year wise)
data.groupby('year')['subject'].value_counts()
```
